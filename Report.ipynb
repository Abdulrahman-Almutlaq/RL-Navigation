## Algorithm used

> This project was implemented using a deep Q-learning (DQN) algorithm, which simply given a state of the enviroment will yield the best action using a neural network, this neural network can be vanilla or CNN (as far as I know).


## Hyperparameters

> The code is from the classroom (with a number of modifications), so I really have not played with the hyperparameters, yet all the values are the traditional, but playing with the hyperparameters is definitely a future work. 

## Architecture

> the architecture of the model consists of 5 fully connected layers
* The first takes 37 (which is the input to the NN, i.e.. number of states), and produce 64
* The second one takes 64 and produces 64
* The third one takes 64 and produces 128
* The fourth one takes 128 and produces 64
* The fifth one takes 64 and produces 4 (which is the number of actions)
